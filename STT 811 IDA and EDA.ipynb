```python
import pandas as pd
import glob
import os

def combine_xlsx_files(input_folder='AI classifier dataset', output_filename='combined.xlsx'):
    print(f"Looking for .xlsx files in folder: {input_folder}")
    full_search_path = os.path.join(input_folder, '*.xlsx')
    print("Full path used:", full_search_path)
    
    excel_files = glob.glob(full_search_path)
    print("Files found:", excel_files)
    
    df_list = []
    
    for file in excel_files:
        try:
            df = pd.read_excel(file)
            df_list.append(df)
        except Exception as e:
            print(f"Error reading {file}: {e}")
    
    if df_list:
        combined_df = pd.concat(df_list, ignore_index=True)
        combined_df.to_excel(output_filename, index=False)
        print(f"Successfully combined {len(excel_files)} files into '{output_filename}'.")
    else:
        print("No valid Excel files found, or an error occurred while reading them.")

if __name__ == "__main__":
    combine_xlsx_files(input_folder='AI classifier dataset', output_filename='combined.xlsx')
```

```python
import pandas as pd
import re

# Load your DataFrame (adjust the filename/path as needed)
df_avg = df_clean

# Define a helper function to count special characters in a string
def count_special_chars(text):
    """
    Count characters that are NOT:
      - Letters (a-z, A-Z)
      - Digits (0-9)
      - Whitespace
    """
    if pd.isnull(text):
        return 0
    # Regex pattern: anything that's NOT a letter, digit, or whitespace
    pattern = r'[^a-zA-Z0-9\s]'
    matches = re.findall(pattern, str(text))
    return len(matches)

# Specify the text columns you want to analyze.
# Adjust these names if your columns are named differently.
text_cols = ['Question', 'Human', 'AI']
# Use only the text columns that actually exist in the DataFrame.
text_cols = [col for col in text_cols if col in df.columns]

# For each text column, create a new column with the special character count
for col in text_cols:
    new_col_name = f"{col}_special_count"
    df_avg[new_col_name] = df_avg[col].astype(str).apply(count_special_chars)

# Function to calculate the average pairwise absolute difference of special char counts for a row.
def avg_pairwise_diff(row, columns):
    # Gather special char counts for the given text columns for this row.
    values = [row[f"{col}_special_count"] for col in columns]
    n = len(values)
    if n < 2:
        return 0  # No difference to compute if only one column exists.
    
    # Compute absolute differences for every unique pair.
    diffs = []
    for i in range(n):
        for j in range(i+1, n):
            diffs.append(abs(values[i] - values[j]))
    
    # Return the average of these differences.
    return sum(diffs) / len(diffs) if diffs else 0

# Create a new column 'avg_special_char_diff' with the computed average difference for each row.
df_avg["avg_special_char_diff"] = df_avg.apply(lambda row: avg_pairwise_diff(row, text_cols), axis=1)

# (Optional) Display the head of the updated DataFrame to verify results.
df_avg.head()
```

Sentimentâ€‘Gap Analysis
Weâ€™re basically measuring mood swings. First, we see how â€œhappyâ€ or â€œgrumpyâ€ the human answer is compared to the mood of the question, and then do the same for the AI answer. We subtract the questionâ€™s mood score from each answerâ€™s mood score. Then we draw two simple pictures:
	â€¢	A bar chart showing how big those mood jumps are for humans vs. the AI.
	â€¢	A scatter plot where each dot shows one questionâ€™s human jump on the bottom and the AI jump on the sideâ€”so you can spot which answers swing more.

```python
from nltk.tokenize import RegexpTokenizer
from gensim.corpora import Dictionary
from gensim.models.ldamodel import LdaModel
import matplotlib.pyplot as plt
import seaborn as sns

# â€”â€”â€”â€”â€”â€”â€”â€”â€”â€”â€”â€”â€”â€”â€”â€”â€”â€”â€”â€”â€”â€”â€”â€”â€”â€”â€”â€”â€”â€”
# 2a. Regexâ€‘based tokenization of each "Question"
# â€”â€”â€”â€”â€”â€”â€”â€”â€”â€”â€”â€”â€”â€”â€”â€”â€”â€”â€”â€”â€”â€”â€”â€”â€”â€”â€”â€”â€”â€”
tokenizer = RegexpTokenizer(r'\w+')  
docs = df['Question'].fillna('').apply(lambda x: tokenizer.tokenize(x.lower()))

# â€”â€”â€”â€”â€”â€”â€”â€”â€”â€”â€”â€”â€”â€”â€”â€”â€”â€”â€”â€”â€”â€”â€”â€”â€”â€”â€”â€”â€”â€”
# 2b. Build gensim Dictionary & Corpus
# â€”â€”â€”â€”â€”â€”â€”â€”â€”â€”â€”â€”â€”â€”â€”â€”â€”â€”â€”â€”â€”â€”â€”â€”â€”â€”â€”â€”â€”â€”
dictionary = Dictionary(docs)
corpus     = [dictionary.doc2bow(doc) for doc in docs]

# â€”â€”â€”â€”â€”â€”â€”â€”â€”â€”â€”â€”â€”â€”â€”â€”â€”â€”â€”â€”â€”â€”â€”â€”â€”â€”â€”â€”â€”â€”
# 2c. Fit LDA (e.g. 5 topics)
# â€”â€”â€”â€”â€”â€”â€”â€”â€”â€”â€”â€”â€”â€”â€”â€”â€”â€”â€”â€”â€”â€”â€”â€”â€”â€”â€”â€”â€”â€”
lda = LdaModel(
    corpus=corpus,
    id2word=dictionary,
    num_topics=5,
    random_state=42,
    passes=10
)

# Print top words per topic
for idx, topic in lda.print_topics(num_words=5):
    print(f"Topic #{idx}: {topic}")

# â€”â€”â€”â€”â€”â€”â€”â€”â€”â€”â€”â€”â€”â€”â€”â€”â€”â€”â€”â€”â€”â€”â€”â€”â€”â€”â€”â€”â€”â€”
# 2d. Assign each question its dominant topic
# â€”â€”â€”â€”â€”â€”â€”â€”â€”â€”â€”â€”â€”â€”â€”â€”â€”â€”â€”â€”â€”â€”â€”â€”â€”â€”â€”â€”â€”â€”
dominant_topics = [
    max(dist, key=lambda x: x[1])[0]
    for dist in lda.get_document_topics(corpus)
]
df['Question_topic'] = dominant_topics

# â€”â€”â€”â€”â€”â€”â€”â€”â€”â€”â€”â€”â€”â€”â€”â€”â€”â€”â€”â€”â€”â€”â€”â€”â€”â€”â€”â€”â€”â€”
# 2e. Visualize topic counts
# â€”â€”â€”â€”â€”â€”â€”â€”â€”â€”â€”â€”â€”â€”â€”â€”â€”â€”â€”â€”â€”â€”â€”â€”â€”â€”â€”â€”â€”â€”
plt.figure(figsize=(6,4))
sns.countplot(x='Question_topic', data=df, palette='tab10')
plt.title('Dominant LDA Topic per Question')
plt.xlabel('Topic')
plt.tight_layout()
plt.show()
```

Topic Modeling (LDA)
Think of this like sorting your questions into five piles based on what words pop up most. We chop each question into its words, build a little â€œword bank,â€ and let the computer figure out five common themes. It then tells you the top five words for each themeâ€”like â€œMarkov,â€ â€œchain,â€ â€œprobability,â€ and so on. Finally, we tag each question with its best-fitting theme and count how many questions fall into each pile.

```python
from nltk.tokenize import RegexpTokenizer
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import textstat

# set up a simple word tokenizer (doesn't need punkt)
tokenizer = RegexpTokenizer(r'\w+')

for col in ['Question','Human','AI']:
    # 1. raw length
    df[f'{col}_char_len'] = df[col].str.len()

    # 2. word counts & unique word counts
    df[f'{col}_word_count'] = df[col].apply(
        lambda x: len(tokenizer.tokenize(x.lower()))
        if isinstance(x, str) else 0
    )
    df[f'{col}_unique_words'] = df[col].apply(
        lambda x: len(set(tokenizer.tokenize(x.lower())))
        if isinstance(x, str) else 0
    )

    # 3. typeâ€“token ratio
    df[f'{col}_ttr'] = (
        df[f'{col}_unique_words'] /
        df[f'{col}_word_count'].replace(0, np.nan)
    )

    # 4. readability scores
    df[f'{col}_flesch'] = df[col].apply(
        lambda x: textstat.flesch_reading_ease(x) if isinstance(x, str) else np.nan
    )
    df[f'{col}_fog'] = df[col].apply(
        lambda x: textstat.gunning_fog(x) if isinstance(x, str) else np.nan
    )

# Example visualization: compare Flesch scores for Human vs AI
plt.figure(figsize=(6,4))
sns.boxplot(
    data=df[['Human_flesch','AI_flesch']].melt(var_name='Source', value_name='Flesch'),
    x='Source', y='Flesch'
)
plt.title('Flesch Reading Ease: Human vs AI')
plt.ylabel('Score')
plt.tight_layout()
plt.show()
```

Textâ€‘Similarity Analysis
This is us checking how close the words are between pairs of texts. We turn each answer into a mathâ€‘style fingerprint of its most important words (TFâ€“IDF). Then for each row, we compare the human vs. AI fingerprints and get a â€œhow similarâ€ number from 0 (total strangers) to 1 (exact match). We do the same comparing the question to each answer. Finally, we plot:
	â€¢	A histogram of humanâ€‘vsâ€‘AI similarity scores, so you see if the AI tends to say things the same way people do.
	â€¢	A smooth curve plot showing how questionâ€‘vsâ€‘human and questionâ€‘vsâ€‘AI scores stack up, so you can see which follows the original questionâ€™s wording more closely.

```python
import networkx as nx
from nltk.tokenize import RegexpTokenizer
from sklearn.feature_extraction.text import ENGLISH_STOP_WORDS
from collections import Counter
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme()

# 1) Combine all Q/A text into one list of strings
texts = df[['Question','Human','AI']].fillna('').agg(' '.join, axis=1).tolist()

# 2) Set up a simple tokenizer + stopâ€‘words
tokenizer  = RegexpTokenizer(r'\w+')
stop_words = set(ENGLISH_STOP_WORDS)

def clean_tokens(doc):
    return [
        tok for tok in tokenizer.tokenize(doc.lower())
        if tok not in stop_words and len(tok) > 2
    ]

# 3) Tokenize every document and flatten into one big list
all_tokens = []
for doc in texts:
    all_tokens += clean_tokens(doc)

# 4) Count bigrams & trigrams
bigrams  = Counter(zip(all_tokens, all_tokens[1:])).most_common(15)
trigrams = Counter(zip(all_tokens, all_tokens[1:], all_tokens[2:])).most_common(15)

# 5) Bar chart helper
def plot_ngrams(ngrams, title):
    labels = [' '.join(gram) for gram,_ in ngrams]
    vals   = [cnt           for _,cnt in ngrams]
    plt.figure(figsize=(8,6))
    plt.barh(labels[::-1], vals[::-1], color='teal')
    plt.title(title)
    plt.xlabel('Frequency')
    plt.tight_layout()
    plt.show()

plot_ngrams(bigrams,  'Top 15 Bigrams')
plot_ngrams(trigrams, 'Top 15 Trigrams')

# 6) Build & draw a bigram network
G = nx.Graph()
for (w1, w2),cnt in bigrams:
    G.add_edge(w1, w2, weight=cnt)

pos     = nx.spring_layout(G, seed=42, k=0.5)
weights = [G[u][v]['weight']*0.1 for u,v in G.edges()]

plt.figure(figsize=(10,8))
nx.draw(G, pos, with_labels=True, width=weights, node_size=300, edge_color='gray')
plt.title('Bigram Coâ€‘occurrence Network')
plt.axis('off')
plt.show()
```
